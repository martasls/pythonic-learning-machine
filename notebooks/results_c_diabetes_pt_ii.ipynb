# Results using SLM-OLS with 20 generations and an optimized learning step and SLM-FLS with 100 generations and learning step equal to 1.

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import os
```

```python
#change working directory
os.chdir("C:\\Users\\Marta\\Documents\\GitHub\\pythonic-learning-machine\\data_sets\\06_formatted\\c_diabetes\\")
```

```python
#testing value 
testing_value = pd.read_csv("testing_value.csv")
testing_value = testing_value[['SLM (FLS)', 'SLM (OLS)', 'SLM (Ensemble)', 'SLM-FLS (Ensemble)', 
                               'SLM (Ensemble-Bagging)', 'SLM-FLS (Ensemble-Bagging)', 'SLM (Ensemble-RIW)',
                               'SLM-FLS (Ensemble-RIW)', 'SLM (Ensemble-Boosting)', 'SLM-FLS (Ensemble-Boosting)']]
testing_value.head(10)
```

```python
fig, ax = plt.subplots()
boxplot = sns.boxplot(data=testing_value, palette="PuBuGn_d")
boxplot.set(xlabel='Algorithms', ylabel='RMSE')
sns.set_context("paper", font_scale=2)
sns.set_style("white")
sns.despine(left=True)
boxplot.set_xticklabels(boxplot.get_xticklabels(), rotation=90)
fig.set_size_inches(18.5, 10.5)
plt.show()
```
